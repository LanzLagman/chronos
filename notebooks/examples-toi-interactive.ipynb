```python
import chronos as cr
```

```python
tois = cr.get_tois()
tois.head()
```

```python
import panel as pn
import hvplot.pandas

x = pn.widgets.Select(name='x', options=tois.columns.tolist())
y = pn.widgets.Select(name='y', options=tois.columns.tolist())
kind = pn.widgets.Select(name='kind', value='scatter', options=['bivariate', 'scatter'])

fields={'x': {'default': 'Period (days)'},
        'y': {'default': 'Planet Radius (R_Earth)'}
       }
plot = tois.hvplot(x=x, y=y, kind=kind, fields=fields, colorbar=False, width=600)
pn.Row(pn.WidgetBox(x, y, kind), plot)
```
